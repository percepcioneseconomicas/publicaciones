```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = 0.5
q = 0.5
B = np.linspace(0.025, 0.1, 100).round(3)
```

```python
N = 100000
n1 = ((N-1)*p*q) / ((N-1)*B**2/4 + p*q)
```

```python
N = 1000000
n2 = ((N-1)*p*q) / ((N-1)*B**2/4 + p*q)
```

```python
N = 10000000
n3 = ((N-1)*p*q) / ((N-1)*B**2/4 + p*q)
```

```python
N = 100000000
n4 = ((N-1)*p*q) / ((N-1)*B**2/4 + p*q)
```

```python
data = pd.DataFrame(
    {
        'N=100.000': n1.astype(int),
        'N=1.000.000': n2.astype(int),
        'N=10.000.000': n3.astype(int),
        'N=100.000.000': n4.astype(int),
    },
    index= B
)
data.loc[0.050]
```

```python
data.plot(figsize=(8,5))
plt.axvline(x=0.05)
plt.axhline(y=399)
plt.grid()
```
