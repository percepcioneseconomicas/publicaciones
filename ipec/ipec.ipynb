```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/percepcioneseconomicas/indices/main/GFK%20ADIMARK%20IPEC.csv', parse_dates=True, dayfirst=True, index_col='Fecha')
data.columns = ['sepa', 'sena', 'senf1', 'senf5', 'cons', 'ipec']
data
```

```python
ax = data['ipec'].plot(label='IPEC', linewidth=1.5, figsize=(8,5))
ax.axhline(50, label='IPEC=50', ls='--', color='green', linewidth=0.8)
ax.axhline(data['ipec'].median(), label='Mediana', ls='--', color='red', linewidth=0.8)
plt. legend()
plt.show()
```

```python
data[['sepa', 'sena', 'cons']].plot(linewidth=0.5, figsize=(8,5))
plt.show()
```
