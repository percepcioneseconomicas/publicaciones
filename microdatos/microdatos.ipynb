```python
import pandas as pd
import matplotlib.pyplot as plt
# plt.style.use('fivethirtyeight')
plt.style.use('Solarize_Light2')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/percepcioneseconomicas/indices/main/UChile%20Indice%20de%20percepcion%20y%20expectativas.csv', parse_dates=True, dayfirst=True, index_col='Fecha')
data.columns = ['Confianza', 'Esperada', 'Actual', 'Familiar', 'País']
data
```

```python
ax = data['Confianza'].plot(label='Confianza', linewidth=1.5, figsize=(8,5))
ax.axhline(data['Confianza'].median(), label='Mediana', ls='--', color='red', linewidth=0.8)
plt. legend()
plt.show()
```

```python
data[['Esperada', 'Actual']].plot(linewidth=1, colormap='coolwarm', figsize=(8,5))
plt.show()
```

```python
data[['Familiar', 'País']].plot(linewidth=1, colormap='coolwarm',figsize=(8,5))
plt.show()
```
